```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
import joblib

# Load dataset
df = pd.read_excel("/content/sample_data/Earnings Manipulator (1).xlsx")

# Define features and target
X = df[["DSRI", "GMI", "AQI", "SGI", "DEPI", "SGAI", "ACCR", "LEVI"]]
y = df["Manipulator"]

# Train-test split
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=42
)

# Train Logistic Regression
logit_model = LogisticRegression(max_iter=1000)
logit_model.fit(X_train, y_train)

# Train Decision Tree (CART)
cart_model = DecisionTreeClassifier(max_depth=4, random_state=42)
cart_model.fit(X_train, y_train)

print("Models trained successfully")
```

```python
joblib.dump(logit_model, "logit_model.pkl")
joblib.dump(cart_model, "cart_model.pkl")

print("Model files saved")
```
